# lume-gpt for Windows

## Preliminary

To use lume-gpt on a Windows machine, first install miniforge (https://github.com/conda-forge/miniforge)

Open a Miniforge Prompt window and install lume-gpt through the instructions in the README as well as Jupyter Lab (https://github.com/jupyterlab/jupyterlab)

## Set up your Windows Powershell Profile ##

Open a Miniforge Prompt window and activate Windows Powershell with the following command:

__powershell__

Along with a notification that powershell launched, your conda environment should now be followed by PS,   ex:  (base) PS

Open your powershell profile for example using notepad with the following command:

__notepad $profile__

Add the following lines to your powershell profile:

----

$env:GPT_BIN = "C:\Program Files\General Particle Tracer\bin\gpt.exe"

$env:GDF2A_BIN = "C:\Program Files\General Particle Tracer\bin\gdf2a.exe"

$env:ASCI2GDF_BIN = "C:\Program Files\General Particle Tracer\bin\asci2gdf.exe"

$env:GPTLICENSE= YOUR_GPT_LICENSE_NUMBER_HERE

----

These lines assume you have gpt installed in the default path "C:\Program files\General Particle Tracer" and should be altered to your GPT installation path.

Replace YOUR_GPT_LICENSE_NUMBER_HERE with your purchased GPT License Number.

The first three lines are 3 commonly used gpt programs for lume-gpt, but this list is not exhaustive.  Other gpt programs can be added using similar syntax.

Changes to your powershell profile will *not* take effect until you open a new miniforge window and reactivate Windows Powershell

## Running lume-gpt ##

Open a new Miniforge Prompt window and activate Windows Powershell with the following command:

__powershell__

Change to your desired directory and launch Jupyter Lab with the following command:

__Jupyter Lab__

lume-gpt should now work, test the example at /lume-gpt/examples/simple_gpt_example.ipynb
